```python
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# import dataset 
dataset = pd.read_csv("covid19_Confirmed_dataset.csv")
dataset.head()
```

```python

# Load datasets
covid_data = pd.read_csv('covid19_Confirmed_dataset.csv')
happiness_data = pd.read_csv('worldwide_happiness_report.csv')

# Display the first few rows of each dataset
print("COVID-19 Data:")
print(covid_data.head())
print("\nHappiness Data:")
print(happiness_data.head())
```

```python
# Check for missing values
print(covid_data.isnull().sum())
print(happiness_data.isnull().sum())
```

```python
# Example: Drop rows with missing values in COVID-19 data
covid_data.dropna(inplace=True)
```

```python
# Remove duplicates
covid_data.drop_duplicates(inplace=True)
happiness_data.drop_duplicates(inplace=True)
```

```python

happiness_data

# Create an area plot
plt.figure(figsize=(10, 6))
happiness_data.plot(kind='area', alpha=0.5)

# Adding titles and labels
plt.title('Area Chart of Happiness Scores and GDP per Capita by Country')
plt.xlabel('Country')
plt.ylabel('Values')

# Adding a legend
plt.legend(title='Metrics')

# Show the plot
plt.tight_layout()
plt.show()

```

```python

happiness_data 
# Create a line chart
plt.figure(figsize=(10, 6))
happiness_data.plot(kind='line', marker='o', linewidth=1)

# Adding titles and labels
plt.title('Line Chart of Happiness Scores and GDP per Capita by Country')
plt.xlabel('Country')
plt.ylabel('Values')

# Adding a legend
plt.legend(title='Metrics')

# Show the plot
plt.grid()
plt.tight_layout()
plt.show()
```
